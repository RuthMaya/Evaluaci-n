```python
#Importamos librerías requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# **EUA**

```python
eua = pd.read_csv("California_EUA.csv", encoding= 'unicode_escape')
```

```python
#Dividir en dos dataframes las columnas numericas y no numericas
datanum = eua.select_dtypes(include = ["number"])
dataobj = eua.select_dtypes(include = ["object"])
```

```python
#VALORES NULOS DE LAS COLUMNAS CATEGORICAS
valores_nulos = dataobj.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores nulos del dataframe con "ffill"
#Se utilizo ffill debido a que este valor permite sustituir las variables con valores nulos por los registros de los datos
#que se encuentran adelante. Esto ayudará a que se conserve la secuencia de los datos y que con la ayuda de valores ya existentes
#en el DF se complementen los registros.
dataobj =dataobj.fillna(method="ffill") 
```

```python
#VALORES NULOS DE LAS COLUMNAS CATEGORICAS
valores_nulos = dataobj.isnull().sum()
valores_nulos
```

```python
#VALORES NULOS DE LAS COLUMNAS NUMERICAS
valores_nulos = datanum.isnull().sum()
valores_nulos
```

```python
#APLICAR PROMEDIO A LAS COLUMNAS CON VALORES NULOS
#Se aplico esta forma ya que esta medida de tendencia central nos permite rellenar los valores con una medida que considera una
#proporción de los datos de la columna, por lo tanto el resto de los valores no se verá afectado.
datanum=datanum.fillna(round(datanum.mean(),1))
```

```python
#VALORES NULOS DE LAS COLUMNAS NUMERICAS
valores_nulos = datanum.isnull().sum()
valores_nulos
```

```python
#VALORES NULOS DE LAS COLUMNAS NUMERICAS
#Se borraron las columnas de los datos que contienen 0's y por lo tanto no son importantes
datanum = datanum.dropna(axis = 1, how='all')
```

```python
#VALORES NULOS DE LAS COLUMNAS NUMERICAS
valores_nulos = datanum.isnull().sum()
valores_nulos
```

### **Eliminar outliers**

```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
#Se aplicó este método debido a que acorta mas los limites entre los datos y nos ayuda para encontrar más outliers.
y=datanum

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Encontramos Outliers del Dataframe
outliers_iqr= datanum[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#Obtenemos datos limpios del Dataframe
data3_iqr= datanum[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data3_iqr
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data3_iqr.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data5_iqr=data3_iqr.copy()
data5_iqr=data5_iqr.fillna(round(data3_iqr.mean(),1))
data5_iqr
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data5_iqr.isnull().sum()
valores_nulos
```

```python
# Unimos los dataframes
Datos_limpios = pd.concat([data5_iqr, dataobj], axis=1)
Datos_limpios
```

```python
Datos_limpios.to_csv("Datos_limpios_EUA.csv")
from google.colab import files 
files.download("Datos_limpios_EUA.csv")
```
