```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from funpymodeling.exploratory import freq_tbl 
```

# **EUA**

```python
eua = pd.read_csv("Datos_limpios_EUA.csv", encoding= 'unicode_escape')
```

```python
#Verificamos información del DataFrame
eua.info()
```

```python
eua.head(3)
```

## **Seleccionar columnas para análisis**

```python
#a - host_response_time
#b - host_acceptance_rate
#c - host_identity_verified
#d - property_type
#e - number_of_reviews
#f - availability_365
#g - review_scores_cleanliness
#h - review_scores_communication
#i - review_scores_location 
#j - review_scores_value
```

### **host_response_time**

```python
#Obtengo un análisis univariado de una variable categórica en específico
table1= freq_tbl(eua['host_response_time'])
table1
```

```python
#Obtengo un filtro de los valores más reelevantes de la variables categórica seleccionada
Filtro= table1[table1['frequency']>1]
Filtro
```

```python
#Ajusto el indice de mi dataframe
Filtro_index= Filtro.set_index('host_response_time')
Filtro_index
```

```python
#Realizamos grafico de pastel del dataframe filtrado
Filtro_index["frequency"].plot(kind='pie', figsize=(10,5), shadow=True, autopct="%0.1f %%")
```

### **host_acceptance_rate**

```python
#Obtengo un análisis univariado de una variable categórica en específico
table1= freq_tbl(eua['host_acceptance_rate'])
table1
```

```python
#Ajusto el indice de mi dataframe
Filtro_index= Filtro.set_index('host_acceptance_rate')
Filtro_index
```

```python
#Realizamos grafico de barras del dataframe filtrado
Filtro_index.plot(kind = 'bar', width=1, figsize=(10,4))
plt.title('host_acceptance_rate')
plt.xlabel('Porcentaje')
plt.ylabel('Frecuencia')
```
